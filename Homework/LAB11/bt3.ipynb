```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('provinces_updated.xlsx')
df['Frequency'] = df['Area'] / df['Population']
df
```

```python
df_sorted = df.sort_values(by='Density', ascending=True).reset_index()
df_sorted
```

```python
# Plot bar chart
plt.figure(figsize=(15, 5))
plt.bar(df_sorted['Density'], df_sorted['Frequency'], color='skyblue')
plt.xlabel('Population Density')
plt.ylabel('Frequency')
plt.title("Population's frequency and density")
plt.tight_layout()
plt.show()
```
