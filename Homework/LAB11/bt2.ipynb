```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('region_areas.xlsx')
df
```

```python
# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Area Proportion (%)'], labels=df["Region"], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Area Proportion by Region')
plt.tight_layout()
plt.show()
```
