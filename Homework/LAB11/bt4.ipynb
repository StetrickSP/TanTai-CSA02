```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('cyclones.csv')
df
```

```python
# Plot line graph
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['cyclone_events'], marker='', linestyle='-', color='b')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Number of Cyclone Events')
plt.title('Cyclone Events Over Time')

# Display the plot
plt.tight_layout()
plt.show()
```
