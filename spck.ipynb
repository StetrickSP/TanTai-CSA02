Sự vươn lên và Phát triển mạnh mẽ của Việt Nam.

```python
### Graph
# Set a list of all years (extracting column names from 1995 to 2023)
years = [str(year) for year in range(1995, 2024)]

# Plot the graph
plt.figure(figsize=(10, 6))

# Loop over each country to plot its line
for index, row in df.iterrows():
    plt.plot(years, row[1:] / 1e9, label=row['Country Name'])

# Add titles and labels
plt.title('GNI Trends of South East Asia countries from 1995 to 2023')
plt.xlabel('Year')
plt.ylabel('GNI (in Billion USD)')
plt.xticks(rotation=45) 

# Add a legend for country names
plt.legend(title="Country Names")

# Display the plot
plt.tight_layout()
plt.show()
```

Nguồn vốn đầu tư từ nước ngoài của Việt Nam, so với lại các cường quốc châu Á.

```python
# Set the years (1990 to 2023) as strings for the x-axis
years = [str(year) for year in range(1990, 2024)]
colors = ["red", "orange", "green", "blue", "purple"]

# Plot the graph
plt.figure(figsize=(10, 6))

# Loop through each country in the DataFrame and plot its values
i = 0
for index, row in df_FDI.iterrows():
    plt.plot(years, row[1:], label=row['Country Name'], color=colors[i])  # Use row[1:] to skip the first column (Country Name)
    i += 1
    
# Add titles and labels
plt.title('FDI of VN compared to other great countries in Asia')
plt.xlabel('Year')
plt.ylabel('Foreign Direct Investment, net inflows (% of GDP)')

# Set x-axis tick marks (rotation to fit all years)
plt.xticks(rotation=45)

# Format the y-axis with major tick marks
plt.gca().yaxis.set_major_formatter(mticker.StrMethodFormatter('{x:,.2f}'))

# Add a legend to identify the countries
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title="Country Names")

# Display the plot with proper layout adjustments
plt.tight_layout()
plt.show()
```

Bảy Quốc gia đầu tư nhiều nhất vào Việt Nam trong năm 2023.

```python
# Set colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']
# Plotting the bar chart
plt.figure(figsize=(8, 5))

# Create a bar chart
plt.bar(df_contributors['Country Name'], df_contributors['USD (in Billion)'], color=colors)

# Add titles and labels
plt.title('Top 7 Contributors of FDI in Viet Nam')
plt.xlabel('Country Names')
plt.ylabel('FDI (USD in Billion)')

# Display the values on top of the bars
for i, value in enumerate(df_contributors['USD (in Billion)']):
    plt.text(i, value + 0.1, f'{value:.2f}', ha='center', va='bottom')

# Show the chart
plt.tight_layout()
plt.show()
```
