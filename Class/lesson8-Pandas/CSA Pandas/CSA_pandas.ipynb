## Hãy tìm ra cầu thủ nào bị nhiều thẻ vàng nhất

```python
lst = [ 9,2,5,7,8,9]

max = 9 
for i in range(len(lst)):
    if lst[i] == max:
        print(lst[i])
```

```python
df[df['YellowCards'] == df['YellowCards'].max()][['Player', 'YellowCards']]
```

```python
# tìm giá trị lớn nhất trong cột thẻ vàng ?
max_yellowcard = df['YellowCards'].max()
max_yellowcard
```

```python
# chạy vòng lặp, search ở từng dòng, xét xem yellowcard của từng cầu thủ có bằng max?
df['YellowCards'] == max_yellowcard

# tạo ra 1 cột mới tên là MaxYellowCards? trong df để lưu trữ giá trị True / False của condition df['YellowCards'] == max_yellowcard
df['MaxYellowCards?'] = df['YellowCards'] == max_yellowcard
```

```python
# lọc df theo 1 condition nào đó
# df[condition True]

df[ df['MaxYellowCards?'] ] [['Player', 'YellowCards']]
```

## Hãy tìm ra cầu thủ ghi nhiều bàn thắng nhất, với vị trí Defender


```python
# logic: 
#     tìm xem ông nào có Position là defender
#     max goals?
df = pd.read_csv('all_players_stats.csv')
df_defender =  df [ df['Position'] == 'Defender' ]
```

```python
df_defender
```

```python
max = df_defender['Goals'].max()

df_defender['Goals'] == max 
```

```python
df_defender [ df_defender['Goals'] == max  ]
```

```python
# hướng 2: tìm tất cả ông có chữ Defender tồn tại trong Position
df = pd.read_csv('all_players_stats.csv')
df_defender2 = df[df['Position'].str.contains('Defender')]
df_defender2
```
